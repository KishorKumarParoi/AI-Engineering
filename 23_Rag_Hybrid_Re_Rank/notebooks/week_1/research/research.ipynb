```python
import pandas as pd
```

```python
import os
from pathlib import Path

# Locate data file
notebook_dir = Path(".").resolve()
data_path = Path(notebook_dir).parent.parent.parent / "data" / "meta_electronics_2026.jsonl"

print(f"Looking for: {data_path}")
print(f"File exists: {data_path.exists()}")

if data_path.exists():
    # Read JSONL with error handling
    records = []
    errors = 0
    
    with open(data_path, 'r') as f:
        for line_num, line in enumerate(f, 1):
            line = line.strip()
            if not line:  # Skip empty lines
                continue
            try:
                record = pd.read_json(line, typ='series')
                records.append(record)
            except Exception as e:
                errors += 1
                if errors <= 3:  # Show first 3 errors
                    print(f"Line {line_num} error: {str(e)[:80]}")
    
    df = pd.DataFrame(records)
    
    # Sample if needed
    if len(df) > 2000:
        df = df.sample(2000, random_state=42)
    
    print(f"\n✓ Loaded {len(df)} valid records ({errors} skipped)")
    print(f"Columns: {df.columns.tolist()[:5]}...")  # Show first 5 columns
else:
    print("✗ Data file not found")
    df = None
```
